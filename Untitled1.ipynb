```python
import matplotlib.pyplot as plt
```

```python
def bresenHum(x0,y0,x1,y1):
    xcoordinate=[]
    ycoordinate=[]

    dx=abs(x0-x1)
    dy=abs(y0-y1)

    p=2*dy-dx

    x=x0
    y=y0

    while x<x1:
        if p>=0:
            y=y+1
            p=p+2*dy-2*dx
        else:
            p=p+2*dy
        x=x+1

        print("x:",x,end=" ")
        print("y:",y,end="\n")

        xcoordinate.append(x)
        ycoordinate.append(y)

    plt.plot(xcoordinate,ycoordinate,marker='o',markersize=3,markerfacecolor="red")
    plt.show()
```

```python
bresenHum(10,15,70,80)
```
