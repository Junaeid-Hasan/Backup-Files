```python
import matplotlib.pyplot as plt
```

```python
def midPointCircle(x_centre,y_centre,radius):
    x=0
    y=radius
    d=1-radius
    plt.plot(x_centre+x,y_centre+y,'ro')
    plt.plot(x_centre+x,y_centre-y,'ro')
    plt.plot(x_centre-x,y_centre+y,'ro')
    plt.plot(x_centre-x,y_centre-y,'ro')

    plt.plot(x_centre+y,y_centre+x,'ro')
    plt.plot(x_centre+y,y_centre-x,'ro')
    plt.plot(x_centre-y,y_centre+x,'ro')
    plt.plot(x_centre-y,y_centre-x,'ro')
    while y>x:
        if d<0:
            d+=2*x+3
        else:
            d+=2*(x-y)+5
            y-=1
        x+=1
        plt.plot(x_centre+x,y_centre+y,'ro')
        plt.plot(x_centre+x,y_centre-y,'ro')
        plt.plot(x_centre-x,y_centre+y,'ro')
        plt.plot(x_centre-x,y_centre-y,'ro')

        plt.plot(x_centre+y,y_centre+x,'ro')
        plt.plot(x_centre+y,y_centre-x,'ro')
        plt.plot(x_centre-y,y_centre+x,'ro')
        plt.plot(x_centre-y,y_centre-x,'ro')

    plt.show()

midPointCircle(3,4,5)
    
```
