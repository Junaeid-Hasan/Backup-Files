```python
import matplotlib.pyplot as plt
```

```python
def DDA(x0,y0,x1,y1):
    dx=abs(x0-x1)
    dy=abs(y0-y1)
    x=float(x0)
    y=float(y0)
    x_coordinate=[]
    y_coordinate=[]

    #dda algo
    max_step=max(dx,dy)
    x_increment=dx/max_step
    y_increment=dy/max_step
    
    #dda for loop
    for i in range(max_step):
        x_coordinate.append(x)
        y_coordinate.append(y)

        x+=x_increment
        y+=y_increment

        print("x:",x,end=" ")
        print("y:",y,end="\n")

    plt.plot(x_coordinate,y_coordinate,marker='o',markersize=3,markerfacecolor="red")
    plt.show()




    
```

```python
DDA(1,5,9,20)
```

```python
def bressenhum(x0,y0,x1,y1):
    dx=abs(x0-x1)
    dy=abs(y0-y1)
    x=x0
    y=y0
    xcoor=[]
    ycoor=[]

    #bressenhum alog
    p=2*dy-dx


    #bressenhum loop
    while x<x1:
        if p>=0:
            y+=1
            p+=2*(dy-dx)
        else:
            p+=2*dy
        x+=1
        xcoor.append(x)
        ycoor.append(y)
        print(f"x:{x},y:{y}")

    plt.plot(xcoor,ycoor,marker='o',markersize=3,markerfacecolor="red")
    plt.show()
```

```python
bressenhum(0,1,7,9)
```
